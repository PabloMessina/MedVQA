```python
from medvqa.evaluation.vqa import aggregate_results_per_question
```

```python
def print_results_per_question(results, dataset_name, metrics_names):
    
    metrics_list = [results[f'{dataset_name}_metrics'][m] for m in metrics_names]
    idxs = results[f'{dataset_name}_metrics']['idxs']
    dataset = results[f'{dataset_name}_dataset']
    tokenizer = results['tokenizer']    
    questions = [tokenizer.ids2string(dataset.questions[i]) for i in idxs]

    aggregate_results_per_question(
        questions=questions,
        metrics_names=metrics_names,
        metrics_list=metrics_list,
    )
```

```python
print_results_per_question(results, 'iuxray', ['bleu_question', 'bleu', 'rougeL', 'ciderD'])
```

```python
print_results_per_question(results, 'mimiccxr', ['bleu_question', 'bleu', 'rougeL', 'ciderD'])
```

```python
from medvqa.evaluation.vqa import plot_vqa_example
```

```python
def inspect_example(results, dataset_name, metrics_names, question, mode):
    
    metrics_list = [results[f'{dataset_name}_metrics'][m] for m in metrics_names]
    dataset = results[f'{dataset_name}_dataset']
    idxs = results[f'{dataset_name}_metrics']['idxs']
    tokenizer = results['tokenizer']
    images = [dataset.images[i] for i in idxs]
    questions = [tokenizer.ids2string(dataset.questions[i]) for i in idxs]
    answers = [tokenizer.ids2string(dataset.answers[i]) for i in idxs]
    pred_questions = [tokenizer.ids2string(x) for x in results[f'{dataset_name}_metrics']['pred_questions']]
    pred_answers = [tokenizer.ids2string(x) for x in results[f'{dataset_name}_metrics']['pred_answers']]    
    
    return plot_vqa_example(question,
                    images,
                    questions,
                    answers,
                    pred_questions,
                    pred_answers,
                    metrics_list,
                    metrics_names,
                    mode = mode)
```

```python
inspect_example(results, 'iuxray', ['bleu_question', 'bleu', 'rougeL', 'ciderD'],
                '<s> what about the trachea </s>', mode='best')
```

```python
inspect_example(results, 'iuxray', ['bleu_question', 'bleu', 'rougeL', 'ciderD'],
                '<s> what about the trachea </s>', mode='worst')
```

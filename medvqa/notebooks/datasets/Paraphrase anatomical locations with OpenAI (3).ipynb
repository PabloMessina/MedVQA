```python
def find_unparaphrased_sentences(
    source_filepath,
    paraphrased_filepaths,
):
    paraphrased_set = set()
    for fp in paraphrased_filepaths:
        paraphrased = load_jsonl(fp)
        for x in paraphrased:
            paraphrased_set.add(x['metadata']['anatomical location'])
            
    source = load_jsonl(source_filepath)
    unparaphrased_set = set()
    for x in source:
        y = x['metadata']['anatomical location']
        if y and y not in paraphrased_set:
            unparaphrased_set.add(y)
    return list(unparaphrased_set)
```

```python
unparaphrased = find_unparaphrased_sentences(
    "/home/pamessina/medvqa-workspace/cache/mimiccxr/integrated_fact_metadata(578718,58628868).jsonl",
    [
        '/home/pamessina/medvqa-workspace/cache/mimiccxr/openai/gpt-3.5-turbo-0613_paraphrased_anatomical_locations__part1.jsonl',
        '/home/pamessina/medvqa-workspace/cache/mimiccxr/openai/gpt-3.5-turbo-0613_paraphrased_anatomical_locations__part2.jsonl',
        '/home/pamessina/medvqa-workspace/cache/mimiccxr/openai/gpt-3.5-turbo-16k-0613_paraphrased_anatomical_locations__part3.jsonl',
    ]
)
```

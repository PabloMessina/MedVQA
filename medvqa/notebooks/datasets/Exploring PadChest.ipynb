```python
def resize_image(img_path, shortest_dim=320):
    orig_img = Image.open(img_path)
    orig_size = img.size
    assert len(orig_size) == 2
    min_dim = min(orig_size)
    new_size = (int(orig_size[0] / min_dim * shortest_dim), int(orig_size[1] / min_dim * shortest_dim))
    new_img = orig_img.resize((new_size))
    new_img.save(img_path)
```

```python
import matplotlib.pyplot as plt

def plot_frequency_per_label(label2count, figsize=(8, 25)):
    pairs = list(label2count.items())
    tot = sum(p[1] for p in pairs)
    n_labels = len(label2count)
    idxs = list(range(n_labels))
    idxs.sort(key=lambda i : pairs[i][1])    
    sorted_counts = [pairs[i][1] for i in idxs]
    sorted_names = [f"{pairs[i][0]} - {pairs[i][1]} ({100 * pairs[i][1] / tot:.2f})%)" for i in idxs]
    
    plt.figure(figsize=figsize)
    plt.barh(list(range(n_labels)), sorted_counts, tick_label=sorted_names)
    plt.show()
```

```python
unique_labels = dict()
for x in df_['Labels'].dropna():
    for y in eval(x):
        y = y.strip()
        unique_labels[y] = unique_labels.get(y, 0) + 1
```

```python
len(unique_labels)
```

```python
plot_frequency_per_label(unique_labels, (7, 60))
```

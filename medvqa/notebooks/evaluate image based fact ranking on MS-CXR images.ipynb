```python
def display_example(examples, i):
    example = examples[i]
    ridx = example['ridx']
    image_path = example['image_path']
    view_pos = example['view_pos']
    rocauc = example['rocauc']
    facts = example['facts']
    fact_probs = example['fact_probs']
    top100_facts = example['top100_facts']
    top100_fact_probs = example['top100_fact_probs']
    
    print_blue(f'ROC AUC: {rocauc:.4f}', bold=True)
    print()
    
    filepath = mimiccxr_metadata['filepaths'][ridx]
    print(filepath)
    print()
    
    print_bold('Report:')
    print(read_txt(filepath))
    print()
    
    print_bold('Image:')
    print(f'image_path: {image_path}')
    print(f'view_pos: {view_pos}')
    image = Image.open(image_path)
    plt.imshow(image)
    plt.axis('off')
    plt.show()

    print_bold('Facts:')
    fact_sorted_idxs = np.argsort(fact_probs)[::-1]
    for idx in fact_sorted_idxs:
        fact = facts[idx]
        fact_prob = fact_probs[idx]
        print(f'{fact_prob:.4f}: {fact}')

    print_bold('Top 100 Facts:')
    for fact, fact_prob in zip(top100_facts, top100_fact_probs):
        if fact in facts:
            print_orange(f'{fact_prob:.4f}: {fact}', bold=True)
        else:
            print(f'{fact_prob:.4f}: {fact}')
```

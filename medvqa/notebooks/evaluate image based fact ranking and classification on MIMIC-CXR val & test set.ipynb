```python
def display_example(examples, i):
    example = examples[i]
    ridx = example['ridx']
    image_path = example['image_path']
    view_pos = example['view_pos']
    rocauc = example['rocauc']
    prcauc = example['prcauc']
    pos_facts = example['pos_facts']
    pos_probs = example['pos_probs']
    neg_facts = example['neg_facts']
    neg_probs = example['neg_probs']
    facts = pos_facts + neg_facts
    fact_probs = np.concatenate([pos_probs, neg_probs])
    
    print_blue(f'ROC AUC: {rocauc:.4f}', bold=True)
    print_blue(f'PRC AUC: {prcauc:.4f}', bold=True)
    print()
    
    print_bold('Image:')
    print(f'image_path: {image_path}')
    print(f'view_pos: {view_pos}')
    image = Image.open(image_path)
    plt.imshow(image)
    plt.axis('off')
    plt.show()
    
    filepath = mimiccxr_metadata['filepaths'][ridx]
    print(filepath)
    print()
    
    print('ridx:', ridx)
    print_bold('Report:')
    print(read_txt(filepath))
    print()

    print_bold('Positive facts:')
    print(pos_facts)
    print()

    print_bold('Ranked facts:')
    fact_sorted_idxs = np.argsort(fact_probs)[::-1]
    for i in fact_sorted_idxs:
        fact = facts[i]
        fact_prob = fact_probs[i]
        if fact in pos_facts:
            print_orange(f'{fact_prob:.4f}: {fact}', bold=True)
        else:
            print(f'{fact_prob:.4f}: {fact}')
```

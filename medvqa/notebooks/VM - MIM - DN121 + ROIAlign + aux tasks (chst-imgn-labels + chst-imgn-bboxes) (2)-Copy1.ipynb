```python
from medvqa.datasets.dataloading_utils import BatchedCompositeInfiniteDataset
from torch.utils.data import Dataset
```

```python
class MyDataset(Dataset):
    def __init__(self, letter, size):
        self.size = size
        self.letter = letter
    def __len__(self):
        return 1e9
    def __getitem__(self, i):
        i = i % self.size
        return f'{self.letter}_{i}'
```

```python
Da = MyDataset('A', 5)
Db = MyDataset('B', 4)
```

```python
bD = BatchedCompositeInfiniteDataset([Da, Db], [1, 1], 10)
```

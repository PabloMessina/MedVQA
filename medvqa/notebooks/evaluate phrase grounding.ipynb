```python
anatomical_locations = set()
for row in integrated_fact_metadata:
    anatloc = row['metadata']['anatomical location']
    if anatloc:
        anatomical_locations.add(anatloc)
anatomical_locations = list(anatomical_locations)
anatomical_locations.sort()
```

```python
len(anatomical_locations)
```

```python
from medvqa.models.huggingface_utils import CachedTextEmbeddingExtractor
```

```python
embedding_extractor = CachedTextEmbeddingExtractor(
    model_name='microsoft/BiomedVLP-CXR-BERT-specialized',
    model_checkpoint_folder_path="/mnt/data/pamessina/workspaces/medvqa-workspace/models/fact_embedding/20231018_230637_MIMIC-CXR(triplets+classif+entcont+nli+autoencoder)_FactEncoder(microsoft-BiomedVLP-CXR-BERT-specialized)",
    num_workers=2,
    batch_size=200,
)
```

```python
anatloc_embeddings = embedding_extractor.compute_text_embeddings(anatomical_locations)
```

```python
anatloc_embeddings.shape
```

```python
from medvqa.utils.math import rank_vectors_by_dot_product
```

```python
query = 'aortic arch'
query_idx = anatomical_locations.index(query)
idxs = rank_vectors_by_dot_product(anatloc_embeddings, anatloc_embeddings[query_idx])

print(query)
print('-' * 50)
for i in range(30):
    print(anatomical_locations[idxs[i]])
```

```python
from medvqa.datasets.nli import RADNLI_DEV_JSONL_PATH, RADNLI_TEST_JSONL_PATH, MS_CXR_T_TEMPORAL_SENTENCE_SIMILARITY_V1_CSV_PATH
radnli_dev_rows = load_jsonl(RADNLI_DEV_JSONL_PATH)
radnli_test_rows = load_jsonl(RADNLI_TEST_JSONL_PATH)
```

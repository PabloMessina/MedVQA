```python
gpt4_radnli_labels = load_jsonl("/home/pamessina/medvqa-workspace/cache/mimiccxr/openai/gpt-4-0613_radnli_queries.jsonl")
```

```python
gpt4_query_to_label = {x['metadata']['query'] : x['parsed_response'] for x in gpt4_radnli_labels}
```

```python
len(gpt4_query_to_label)
```

```python
from medvqa.datasets.nli import RADNLI_DEV_JSONL_PATH, RADNLI_TEST_JSONL_PATH
radnli_dev_rows = load_jsonl(RADNLI_DEV_JSONL_PATH)
radnli_test_rows = load_jsonl(RADNLI_TEST_JSONL_PATH)
```

```python
dev_acc = 0
for row in radnli_dev_rows:
    query = f'Premise: {row["sentence1"]} | Hypothesis: {row["sentence2"]}'
    if gpt4_query_to_label[query] == row['gold_label']:
        dev_acc += 1
dev_acc /= len(radnli_dev_rows)
print('dev_acc:', dev_acc)
```

```python
test_acc = 0
mistakes = []
gt_labels = []
pred_labels = []
label_names = ['entailment', 'neutral', 'contradiction']
for row in radnli_test_rows:
    query = f'Premise: {row["sentence1"]} | Hypothesis: {row["sentence2"]}'
    gt_labels.append(label_names.index(row['gold_label']))
    pred_labels.append(label_names.index(gpt4_query_to_label[query]))
    if gpt4_query_to_label[query] == row['gold_label']:
        test_acc += 1
    else:
        mistakes.append({
            'query': query,
            'premise': row["sentence1"],
            'hypothesis': row["sentence2"],
            'gpt-4': gpt4_query_to_label[query],
            'gold_label': row['gold_label'],
        })
test_acc /= len(radnli_test_rows)
print('test_acc:', test_acc)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
cm = confusion_matrix(gt_labels, pred_labels)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=label_names, yticklabels=label_names)
# sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

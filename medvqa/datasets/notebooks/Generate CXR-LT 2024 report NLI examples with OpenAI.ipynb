```python
def show_report(study_id):
    idx = metadata['study_ids'].index(study_id)
    return _build_report(metadata, idx)
```

```python
import random
# class_name = 'Scoliosis'
class_name = 'Hilum'
source_name = 'ChestImageNome'
# source_name = 'Regex'
# source_name = 'CXRLT2024'

for key, val in tmp[class_name].items():
    print (key, len(val))

study_id = random.choice(tmp[class_name][source_name])
show_report(study_id)
```

```python
import re
n_reports = len(metadata['backgrounds'])
reports = [_build_report(metadata, i) for i in range(n_reports)]
regex = re.compile(r'\beffusion', re.IGNORECASE)
study_ids = []
for study_id, report in zip(metadata['study_ids'], reports):
    if regex.search(report):
        study_ids.append(study_id)
assert len(study_ids) > 0
```

```python
unique_facts = set()
for x in integrated_sentence_facts:
    unique_facts.update(x['facts'])
```

```python
len(unique_facts)
```

```python
unique_facts_list = list(unique_facts)
```

```python
from nltk.tokenize import word_tokenize
from tqdm import tqdm

word_count = dict()
tokenized_facts = [None] * len(unique_facts_list)
for i, fact in tqdm(enumerate(unique_facts_list), mininterval=2):
    words = word_tokenize(fact)
    tokenized_facts[i] = words
    for w in words:
        word_count[w] = word_count.get(w, 0) + 1
```

```python
def _difficulty(i):
    score = 0
    for w in tokenized_facts[i]:
        score += 1/word_count[w]
    return score

ranked_indices = sorted(range(len(unique_facts_list)), key=_difficulty, reverse=True)
```

```python
query = 'A BENIGN CALCIFICATION IN LUNG OR RIB, OR VESSEL, EITHER ON END OR CROSSING in the right lower lung'
for x in integrated_sentence_facts:
    for f in x['facts']:
        if f == query:
            print(x)
```

```python
from medvqa.scripts.mimiccxr.generate_classification_labels_with_openai import VINBIG_LABEL_TO_HYPOTHESIS
h2c = { h:c for c,h in VINBIG_LABEL_TO_HYPOTHESIS.items() }
class2counts = {}
for x in rows:
    q = x['metadata']['query']
    l = x['parsed_response']['label']
    h = q[q.index(' | #H: ') + 7:]
    c = h2c[h]
    try:
        counts = class2counts[c]
    except KeyError:
        counts = {}
        class2counts[c] = counts
    try:
        counts[l] += 1
    except KeyError:
        counts[l] = 1
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = class2counts  # Paste your dictionary here

categories = ["certainly false", "probably false", "unknown", "probably true", "certainly true"]
colors = ["#d62728", "#ff7f0e", "#bcbd22", "#1f77b4", "#2ca02c"]  # Red, Orange, Yellow, Blue, Green

# Extract keys (conditions) and values (category counts)
conditions = list(data.keys())
values = np.array([[data[cond].get(cat, 0) for cat in categories] for cond in conditions])

# Compute positions and total counts
x = np.arange(len(conditions))
totals = values.sum(axis=1)

# Sort by totals
idxs = totals.argsort()[::-1]
totals = totals[idxs]
values = values[idxs]
conditions = [conditions[i] for i in idxs]

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Plot stacked bars
bottom = np.zeros(len(conditions))
for i, cat in enumerate(categories):
    ax.bar(x, values[:, i], label=cat, color=colors[i], bottom=bottom)
    bottom += values[:, i]

# Add total count above bars
for i, total in enumerate(totals):
    ax.text(x[i], total + max(totals) * 0.02, str(total), ha='center', fontsize=9, fontweight='bold')

# Formatting
ax.set_xticks(x)
ax.set_xticklabels(conditions, rotation=90, fontsize=8)
ax.set_ylabel("Count")
ax.set_title("Condition Distribution by Certainty Level")
ax.legend(title="Certainty Level")
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Extract label names
label_names = val_dict['label_names']

# Get the number of labels
num_classes = len(label_names)

# Count positives and negatives per class
val_labels = np.array(list(val_dict['dicom_id_to_labels'].values()))
test_labels = np.array(list(test_dict['dicom_id_to_labels'].values()))

val_positives = np.sum(val_labels, axis=0)
val_negatives = val_labels.shape[0] - val_positives

test_positives = np.sum(test_labels, axis=0)
test_negatives = test_labels.shape[0] - test_positives

# Compute sorting indices based on the average number of positives
avg_positives = (val_positives + test_positives) / 2
sorted_indices = np.argsort(avg_positives)[::-1]  # Sort descending

# Apply sorting
label_names = [label_names[i] for i in sorted_indices]
val_positives = val_positives[sorted_indices]
val_negatives = val_negatives[sorted_indices]
test_positives = test_positives[sorted_indices]
test_negatives = test_negatives[sorted_indices]

# Plot settings
x = np.arange(num_classes)
width = 0.4  # Width of bars

fig, ax = plt.subplots(figsize=(12, 6))

# Validation bars
val_bars = ax.bar(x - width/2, val_positives, width, label="Val Positive", color="tab:blue", alpha=0.7)
ax.bar(x - width/2, val_negatives, width, bottom=val_positives, label="Val Negative", color="tab:blue", alpha=0.3)

# Test bars
test_bars = ax.bar(x + width/2, test_positives, width, label="Test Positive", color="tab:orange", alpha=0.7)
ax.bar(x + width/2, test_negatives, width, bottom=test_positives, label="Test Negative", color="tab:orange", alpha=0.3)

# Add text labels above bars
for bars, positives in zip([val_bars, test_bars], [val_positives, test_positives]):
    for bar, pos in zip(bars, positives):
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 50, str(int(pos)),
                ha='center', va='bottom', fontsize=9, fontweight='bold')

# Labels and formatting
ax.set_xlabel("Classes")
ax.set_ylabel("Count")
ax.set_title("Class Distribution in Validation and Test Sets (Sorted by Positives)")
ax.set_xticks(x)
ax.set_xticklabels(label_names, rotation=90)
ax.legend()

plt.tight_layout()
plt.show()
```

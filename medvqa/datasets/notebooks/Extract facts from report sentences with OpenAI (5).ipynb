```python
def _compute_gpt4_sentence_to_facts(gpt4_extracted_facts_filepaths):
    s2fs = {}
    for fp in gpt4_extracted_facts_filepaths:
        rows = load_jsonl(fp)
        for row in rows:
            m = row['metadata']
            try:
                s = m['query']
            except KeyError:
                s = m['sentence']
            fs = row['parsed_response']
            s2fs[s] = fs
    return s2fs
```

```python
s2fs = _compute_gpt4_sentence_to_facts([
    "/home/pamessina/medvqa-workspace/cache/mimiccxr/openai/gpt-4-0613_parsed_sentences__v2.jsonl",
    "/home/pamessina/medvqa-workspace/cache/mimiccxr/openai/gpt-4-0613_facts_from_sentences(cluster-balanced,hardest,conditional).jsonl",
    "/home/pamessina/medvqa-workspace/cache/mimiccxr/openai/gpt-4-0613_facts_from_sentences(mimiccxr_test_set).jsonl"
])
```
